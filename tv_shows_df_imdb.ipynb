```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.imdb.com/chart/toptv/?ref_=nv_tvv_250"
page = requests.get(url)
soup = BeautifulSoup(page.text, "html.parser")
```

```python
table = soup.find_all("table")[0]
rank_rows = table.find_all("span", {"name": "rk"})
title_rows = table.find_all("td", {"class": "titleColumn"})
actors_rows = table.find_all("td", {"class": "titleColumn"})
release_year_rows = table.find_all("span", {"class": "secondaryInfo"})
rating_rows = table.find_all("strong")

rank = [x.attrs["data-value"] for x in rank_rows]
title = [x.a.text for x in title_rows]
actors = [x.a.attrs["title"] for x in actors_rows]
release_year = [x.text[1:-1] for x in release_year_rows]
rating = [x.text for x in rating_rows]

pd.DataFrame({"Rank" :rank, "Title": title, "Actors": actors, "Release Year": release_year, "Rating": rating})
```
